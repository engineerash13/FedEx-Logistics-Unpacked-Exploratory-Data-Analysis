## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

df1=df.copy()
````

````python
# 1. Handle Missing Values

# Drop rows with missing 'Dosage' since it has the highest % of missing values (16.8%)
df1 = df1.dropna(subset=['Dosage'])

# Fill missing 'Shipment Mode' with the mode (most frequent value)
df1['Shipment Mode'].fillna(df1['Shipment Mode'].mode()[0], inplace=True)

# Fill missing 'Line Item Insurance (USD)' with 0 or mean/median
df1['Line Item Insurance (USD)'].fillna(0, inplace=True)
````

````python
# 2. Convert Columns to Correct Data Types

# Remove commas/dollar signs if present and convert to float
df1['Weight (Kilograms)'] = pd.to_numeric(df1['Weight (Kilograms)'].str.replace(',', ''), errors='coerce')
df1['Freight Cost (USD)'] = pd.to_numeric(df1['Freight Cost (USD)'].str.replace(',', ''), errors='coerce')
````

````python
# 3. Convert Date Columns to datetime

date_cols = [
    'Delivery Recorded Date',
    'Delivered to Client Date',
    'Scheduled Delivery Date',
    'PO Sent to Vendor Date',
    'PQ First Sent to Client Date'
]

for col in date_cols:
    df1[col] = pd.to_datetime(df1[col], errors='coerce')
````

````python
# 4. Create New Useful Columns

# Example: Calculate delivery delay
df1['Delivery Delay (Days)'] = (df1['Delivered to Client Date'] - df1['Scheduled Delivery Date']).dt.days
````

````python
 # 5. Drop Duplicates

 df1.drop_duplicates(inplace=True)
````

````python
# 6. Reset Index (after dropping rows)

df1.reset_index(drop=True, inplace=True)
````

#### Chart - 7 Monthly Delivery Trend - On Time vs Late (Line Chart)

````python
# Convert date columns to datetime
df1['Scheduled Delivery Date'] = pd.to_datetime(df1['Scheduled Delivery Date'], errors='coerce')
df1['Delivered to Client Date'] = pd.to_datetime(df1['Delivered to Client Date'], errors='coerce')

# Derive Delivery Status
df1['Delivery Status'] = df1['Delivered to Client Date'] <= df1['Scheduled Delivery Date']
df1['Month'] = df1['Delivered to Client Date'].dt.to_period('M').astype(str)

# Count deliveries by month and status
monthly_trend = df1.groupby(['Month', 'Delivery Status']).size().reset_index(name='Count')

# Plotting
fig = px.line(
    monthly_trend,
    x='Month',
    y='Count',
    color='Delivery Status',
    markers=True,
    title='Monthly Delivery Trend - On Time vs Late',
    color_discrete_map={True: 'green', False: 'red'}
)

fig.update_layout(
    xaxis_title='Month',
    yaxis_title='Number of Deliveries',
    legend_title='On Time'
)

fig.show()

````

##### 1. Why did you pick the specific chart?

I chose a line chart to visualize the trend of delivery performance over time, specifically distinguishing between on-time and late deliveries on a monthly basis. Line charts are ideal for time-series data, allowing us to observe fluctuations, patterns, and seasonality in delivery behavior across years.

#### Chart - 9 Average Delivery Delay by Vendor and Product Group (Density Heatmap)

````python
# Chart - 9 visualization code

# Group by Vendor and Product Group and calculate the mean delay
heatmap_data = df1.groupby(['Vendor', 'Product Group'])['Delivery Delay (Days)'].mean().reset_index()

# Create a heatmap using Plotly
fig = px.density_heatmap(
    heatmap_data,
    x='Product Group',
    y='Vendor',
    z='Delivery Delay (Days)',
    color_continuous_scale='Reds',
    title='Average Delivery Delay by Vendor and Product Group'
)

# Update axis labels
fig.update_layout(
    xaxis_title='Product Group',
    yaxis_title='Vendor'
)

# Show the heatmap
fig.show()
````

##### 1. Why did you pick the specific chart?

A heatmap is ideal for visualizing patterns and comparing two categorical variables (Vendor & Product Group) against a quantitative variable (Average Delivery Delay).
This chart makes it very easy to spot vendors and product categories where delays are higher or lower, using intuitive color intensity.

##### 2. What is/are the insight(s) found from the chart?
